```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

Reading Covid Data

```python
covid_dataset = pd.read_csv("C:/Senthan/Python/pandas/data/covid_data/covid_data.csv")
print(covid_dataset)
```

Making new columns

```python
covid_dataset["Recovered+Active"] = covid_dataset["Recovered"]+covid_dataset["Active"]
covid_dataset["Cases"] = covid_dataset["Confirmed"]
covid_dataset_cases = covid_dataset.nlargest(20, "Confirmed")
print(covid_dataset_cases)
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(data=covid_dataset_cases, x="Country/Region", y="Cases", color="red")
sns.barplot(data=covid_dataset_cases, x="Country/Region", y="Recovered+Active", color="blue")
sns.barplot(data=covid_dataset_cases, x="Country/Region", y="Recovered", color="green")
plt.xticks(rotation = 90, fontsize = 6)

label1 = plt.bar([0], [0], color="red")
label2 = plt.bar([0], [0], color="blue")
label3 = plt.bar([0], [0], color="green")
plt.legend(handles=[label1[0], label2[0], label3[0]],labels=["Dead", "Active", "Recovered"], title="Cases")
plt.title("Covid Cases by Top 20 Countries")
plt.show()
```

```python
def change_percent_to_num(percent):
    if percent<=1:
        percent = percent*100
    return percent
```

```python
covid_dataset["Total Percentage"] = 100
covid_dataset["Recovered / 100 Cases"] = covid_dataset["Recovered / 100 Cases"].map(change_percent_to_num)

covid_dataset_1 = covid_dataset.nlargest(20, "Confirmed")
print(covid_dataset_1)
```

```python
sns.set_theme(style="ticks")
sns.barplot(data=covid_dataset_1, x="Country/Region", y="Total Percentage", color="red")
sns.barplot(data=covid_dataset_1, x="Country/Region", y="Recovered / 100 Cases", color="green")
plt.xticks(rotation = 90, fontsize = 6)
# tick_labels = plt.gca().get_xticklabels()
# for i, label in enumerate(tick_labels):
#     if i % 5 != 0:
#         label.set_visible(False)


label1 = plt.bar([0], [0], color="red")
label3 = plt.bar([0], [0], color="green")
plt.legend(handles=[label1[0], label3[0]],labels=["Dead","Recovered"], title="Cases")
plt.title("Covid Cases by Top 20 Countries")
plt.show()
```
